```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    
    return True

# Function is checking by coverting x into float and if it is not a number it is range or it is different then
# it will go on except block and return false othervise return true
```

```python
data3[~data3['total_sqft'].apply(is_float)].head(20)
# Above state ment will return all true value for seeing the value with range or different errors do negation
```

```python
def total_sqft_converter(x):
    a = x.split('-')
    if len(a) == 2:
        return ((float(a[0])+float(a[1]))/2)
    try:
        return float(x)
    except:
        return None
```

```python
data4 = data3.copy()
data4['total_sqft'] = data3['total_sqft'].apply(total_sqft_converter)
data4.head(10)
```

```python
data4['price_per_sqft'] = data4['price']*100000/data4['total_sqft']
# As price is in lakh multiply it with 100000
data4
```

```python
data5 = data4[~((data4.total_sqft/data4.BHK)<300)]
data5.shape
# So we removed the one outlier of flat which has less than 300 sqft BHK
```

```python
import matplotlib.pyplot as plt
# Now we will check whether the 2BHK price is less than 3BHK price of same location and sqft

def plot_scatter_chart(data,location):
    bhk2 = data[(data.location == location) & (data.BHK == 2)]
    bhk3 = data[(data.location == location) & (data.BHK == 3)]
    plt.rcParams['figure.figsize'] = (15,10)
    plt.scatter(bhk2.total_sqft,bhk2.price,color='blue',label='2 BHK',s=50)
    plt.scatter(bhk3.total_sqft,bhk3.price,marker='+',color='green',label="3 BHK",s=50)
    plt.xlabel("Total Square Feet")
    plt.ylabel("Price")
    plt.legend()
    
plot_scatter_chart(data5,"Hebbal")

# As you can see below at 1400 total_squareft price of 3BHK apartment is less than 2 BHK for same sqft
```

```python
# We Should also remove properties where for same location the price (of example) 3 bedroom apartment is less than 2 bedroom 
# apartment (with same square feet area). What we will do for given location we will build of stats per bhk i.e
# {
#     '1' : {
#         'mean' : 4000;
#         'std' : 2000;
#         'count' : 34;
#     },
    
#     '2' : {
#         'mean' : 4300;
#         'std' : 2300;
#         'count' : 22;
#     },
# }

# Now we can remove 2 BHK apartments whose price_per_sqft is less than mean price_per_sqft of 1 BHK apartment
```

```python
def remove_bhk_outliers(data):
    exclude_data = np.array([])
    
    for location, location_data in data.groupby('location'):
        bhk_stats = {}
        for BHK,BHK_data in location_data.groupby('BHK'):
            bhk_stats[BHK] = {
                'mean' : np.mean(BHK_data.price_per_sqft),
                'std' : np.std(BHK_data.price_per_sqft),
                'count': BHK_data.shape[0],
            }
        for BHK, BHK_data in location_data.groupby('BHK'):
            stats = bhk_stats.get(BHK-1)
            if stats and stats['count'] > 5:
                exclude_data = np.append(exclude_data,BHK_data[BHK_data.price_per_sqft<(stats['mean'])].index.values)
    
    return data.drop(exclude_data,axis='index')

data6 = remove_bhk_outliers(data5)
data6.shape
```

```python
data8 = data7[data7.bath<data7.BHK+2]
data8
```

```python
data9 = data8.drop(['size','price_per_sqft'],axis=1)
data9.head(10)
```

```python
# For Converting string data into numerical in panda we have get_dummies
# dummies converts the col into 1 and 0 i.e if first there is location Chikka Tirupati then the value of that will be one
#  and others location value will be zero and same for others location also
# For eg above at 12th location is 7th Phase JP Nagar so that location will be 1 and others will be 0
dummies = pd.get_dummies(data9.location)
dummies.head(10)
```

```python
data10 = pd.concat([data9,dummies],axis=1)
data10.head(10)
```

```python
# Now we will drop location as it is in string and we have make a dummies of location column
data11 = data10.drop('location',axis=1)
data11
```

```python
data11.shape
```

```python
# We will gonna drop price in X bcoz we just want InDependent values
X = data11.drop('price',axis=1)
X
```

```python
Y = data11.price
Y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=10)
# test_size specify that 30% of my data will be for test and remaining for training
```

```python
# First we will try the Linear Regression and check the score if the score will be nice then proceed further othervise,
# we will try with another till we get Nice score

from sklearn.linear_model import LinearRegression

lr = LinearRegression(normalize=True)
lr.fit(X_train,Y_train)
lr.score(X_test,Y_test)
```

```python
# Lets try by using K fold cross validation

from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5,test_size=0.2,random_state=0)

#ShuffleSplit will randomize the sample so that each of the fold will have equal number of data samples in each iteration

cross_val_score(LinearRegression() , X, Y,cv=cv)
```

```python
def price_predict(location,sqft,bath,bhk):
    
    loc_index = np.where(X.columns == location)[0][0]
    
    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    
    if loc_index >=0:
        x[loc_index] = 1
    
    return lr.predict([x])[0]
# predict function gives us the array so are fetching the 1st position inside of array
```
